```python
import geopandas as gpd
```

```python
gdf = gpd.read_file("./cell_coverage.csv")
gdf.head()
```

```python
gdf.columns
```

```python
gdf = gdf.drop(columns = ['Scr_Freq', 'UARFCN', 'Bearing', 'TiltE', 'TiltM', 'SiteID', 'City',
       'Height', 'TAC', 'Band', 'Vendor', 'MaxTransPwr', 'FreqMHz', 'HBW', 'RF_Team', 'max_dist_1_dt',
       'perc_dist_reduct_1_dt', 'max_dist_2_dt', 'perc_dist_reduct_2_dt',
       'max_dist_1_ut', 'perc_dist_inc_1_ut', 'max_dist_2_ut', 'perc_dist_inc_2_ut', 'avg_rsrp_1_degree_downtilt',
       'avg_rsrp_2_degree_downtilt', 'avg_rsrp_1_degree_uptilt', 'avg_rsrp_2_degree_uptilt'])
```

```python
gdf.rename(columns = {'Latitude' : 'latitude', 'Longitude' : 'longitude'}, inplace = True)
```

```python
gdf['band_cell_count'] = (
    gdf
    .groupby(['grid', 'Band'])['cell_name']
    .transform('count')
)
```

```python
gdf['cell_count'] = (
    gdf
    .groupby(['grid'])['cell_name']
    .transform('count')
)
```

```python
gdf.to_csv("./cell_coverage.csv", index = False)
```
