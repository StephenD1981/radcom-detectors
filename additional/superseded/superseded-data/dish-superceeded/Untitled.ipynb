```python
# Earth radius in km
R = 6371.0

lat1 = np.radians(impacts_denv['cell_lat'])
lon1 = np.radians(impacts_denv['cell_lon'])
lat2 = np.radians(impacts_denv['cell_impact_lat'])
lon2 = np.radians(impacts_denv['cell_impact_lon'])

dlat = lat2 - lat1
dlon = lon2 - lon1

a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

impacts_denv['distance'] = R * c * 1000
```
