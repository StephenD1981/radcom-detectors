```python
import geohash
from shapely.geometry import Polygon

def geohash_polygon(gh: str) -> Polygon | None:
    if gh is None:
        return None
    try:
        lat, lon, lat_err, lon_err = geohash.decode_exactly(gh)
        lat_min, lat_max = lat - lat_err, lat + lat_err
        lon_min, lon_max = lon - lon_err, lon + lon_err

        # Polygon coordinates are (lon, lat)
        return Polygon([
            (lon_min, lat_min),
            (lon_min, lat_max),
            (lon_max, lat_max),
            (lon_max, lat_min),
            (lon_min, lat_min)
        ])
    except Exception:
        return None

gdf["geohash_poly"] = gdf["grid"].map(geohash_polygon)
```

```python
final_gdf = gdf.rename(columns = {"geohash_poly": "geometry"})
```
