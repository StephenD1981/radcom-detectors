```python
import pandas as pd
pd.set_option('display.max_columns', None)
```

```python
impacts = pd.read_csv("./impact_matrix.csv")
gis = pd.read_csv("./cork-gis.csv")
```

```python
def left(s, amount):
    return s[:amount]
```

```python
print("Pre: ", impacts.shape)
impacts_filtered = impacts[(impacts['cell_name'].str[:2] == "CK") & (impacts['cell_impact_name'].str[:2] == "CK")]
print("Post: ", impacts_filtered.shape)
```

```python
impacts_filtered = impacts_filtered.rename(columns = {'Cell' : 'cell', 'Site' : 'site', 'cell_Tech' : 'cell_tech', 
                            'LTE_on_cell' : 'lte_on_cell', 'Impact_Cell' : 'impact_cell', 'Impact_Site' : 'impact_site',
                            'cell_impact_Tech' : 'cell_impact_tech', 'LTE_on_impact_cell' : 'lte_on_impact_cell', 
                            'Co-Sectored' : 'co_sectored', 'Co-Site' : 'co_site', 'Traffic_Data' : 'traffic_data',
                            'Total_Cell_Traffic_Data' : 'total_cell_traffic_data', 
                            'Total_Cell_Traffic_Data_LTE' : 'total_cell_traffic_data_lte', 
                            'total_LTE_impact_data(%)' : 'total_lte_impact_data(%)', 
                            'Total_Cell_Traffic_Data_UMTS' : 'total_cell_traffic_data_umts',
                            'total_UMTS_impact_data(%)' : 'total_umts_impact_data(%)', 
                            'Total_Cell_Traffic_Data_GSM' : 'total_cell_traffic_data_gsm',
                            'total_GSM_impact_data(%)' : 'total_gsm_impact_data(%)',
                            'Traffic_Voice' : 'traffic_voice', 'Total_Cell_Traffic_Voice' : 'total_cell_traffic_voice',
                            'Total_Cell_Traffic_Voice_LTE' : 'total_cell_traffic_voice_lte',
                            'Total_Cell_Traffic_Voice_UMTS' : 'total_cell_traffic_voice_umts',
                            'total_UMTS_impact_voice(%)' : 'total_umts_impact_voice(%)',
                            'Total_Cell_Traffic_Voice_GSM' : 'total_cell_traffic_voice_gsm',
                            'total_GSM_impact_voice(%)' : 'total_gsm_impact_voice(%)', 'Drops_Voice' : 'drops_voice',
                            'EventsOfImpactDurationOnCellB' : 'events_of_impact_duration_on_cell_b',
                            'impact_site_requires_LTE' : 'impact_site_requires_lte',
                            'relation_impact_data(%)' : 'relation_impact_data_perc',
                            'total_lte_impact_data(%)' : 'total_lte_impact_data_perc',
                            'total_UMTS_impact_data(%)' : 'total_UMTS_impact_data_perc',
                            'total_GSM_impact_data(%)' : 'total_gsm_impact_data_perc',
                            'relation_impact_voice(%)' : 'relation_impact_voice_perc',
                            'total_LTE_impact_voice(%)' : 'total_LTE_impact_voice_perc',
                            'total_umts_impact_voice(%)' : 'total_umts_impact_voice_perc',
                            'total_gsm_impact_voice(%)' : 'total_gsm_impact_voice_perc',
                            'total_lte_impact_data(%)' : 'total_lte_impact_data_perc',
                            'total_umts_impact_voice(%)' : 'total_umts_impact_voice_perc',
                            'total_gsm_impact_voice(%)' : 'total_gsm_impact_voice_perc',
                            'co-sector_downgrades' : 'co_sector_downgrades',
                            'co-sector_downgrades_action' : 'co_sector_downgrades_action'})
impacts_filtered.columns
```

```python
impacts_filtered.head(2)
```

```python
lte_site_list = list(set(impacts_filtered[impacts_filtered["cell_tech"] == "LTE"]["site"]))
lte_impact_site_list = list(set(impacts_filtered[impacts_filtered["cell_impact_tech"] == "LTE"]["impact_site"]))
gsm_site_list = list(set(impacts_filtered[impacts_filtered["cell_tech"] == "GSM"]["site"]))
gsm_impact_site_list = list(set(impacts_filtered[impacts_filtered["cell_impact_tech"] == "GSM"]["impact_site"]))
umts_site_list = list(set(impacts_filtered[impacts_filtered["cell_tech"] == "UMTS"]["site"]))
umts_impact_site_list = list(set(impacts_filtered[impacts_filtered["cell_impact_tech"] == "UMTS"]["impact_site"]))
print(len(lte_site_list), len(lte_impact_site_list), len(gsm_site_list),
      len(gsm_impact_site_list), len(umts_site_list), len(umts_impact_site_list))
```

```python
impacts_filtered["lte_on_cell"] = ["Y" if j in lte_site_list else "N" for j in impacts_filtered['site']]
impacts_filtered["lte_on_impact_cell"] = ["Y" if j in lte_impact_site_list else "N" for j in impacts_filtered['impact_site']]
impacts_filtered["gsm_on_cell"] = ["Y" if j in gsm_site_list else "N" for j in impacts_filtered['site']]
impacts_filtered["gsm_on_impact_cell"] = ["Y" if j in gsm_impact_site_list else "N" for j in impacts_filtered['impact_site']]
impacts_filtered["umts_on_cell"] = ["Y" if j in umts_site_list else "N" for j in impacts_filtered['site']]
impacts_filtered["umts_on_impact_cell"] = ["Y" if j in umts_impact_site_list else "N" for j in impacts_filtered['impact_site']]
```

```python
impacts_filtered["nsa_on_cell"] = "N"
impacts_filtered["nsa_on_impact_cell"] = "N"
impacts_filtered["sa_on_cell"] = "N"
impacts_filtered["sa_on_impact_cell"] = "N"
```

```python
impacts_filtered.to_csv("./cork_impacts.csv", index = False)
```
